```python
import pandas as pd

df = pd.read_csv("titanic.csv")

print ("Vista previa de los datos del Titanic:")
print (df.head())


#Análisis de supervivencia

porcentaje_sobreviviente = df["Survived"].mean() *100
print(f"\nPorcentaje de sobrevivientes: {porcentaje_sobreviviente:.2f} %")

#Análisis de Clase

porcentaje_clase = df["Parch"].value_counts(normalize=True) *100
print (f"\nResultado de análisis de clase (Parch):\n {porcentaje_clase}")

#Análisis por Genero

genero = df["Sex"].value_counts()
print (f"\n Resultado por género: \n{genero}")

```

#Coclusiones


1.   En síntesis, el porcentaje de supervivencia del análisis de datos del Titanic, se observa que sobrevivieron menos de la mitad del porcentaje total, lo que demostró el tamaño del desastre y la escasa capacidad de evacuación del barco.
2.   Se deduce que, a través del análisis de género y clase, influye de manera significativa tanto la clase social como el género en las posibilidades de supervivencia, considerando que la mayoría de los pasajeros viajaban solos, además que los datos obtenidos muestran que los hombres son  los que más viajaban que las mujeres, sin embargo, las mujeres tuvieron mayor porcentaje de sobrevivientes.
